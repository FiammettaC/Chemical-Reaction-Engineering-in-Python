## Problem A: Ideal Batch Reactor
In a constant volume batch reactor (due to constant density), the mole balance simplifies to:
$$\frac{dC_A}{dt} = r_A$$[cite: 2]

### 1. Rate Constant Calculation
We are given that at $t = 1$ hour, the concentration ratio is $\frac{C_A}{C_{A0}} = 0.1$. We need to find how much longer it takes to reach $\frac{C_A}{C_{A0}} = 0.001$.

```python
import numpy as np

# Given constants
t_1 = 1.0  # hour
ratio_1 = 0.1
ratio_target = 0.001

# --- Case a: First Order (n=1) ---
# ln(CA/CA0) = -k*t => k = -ln(ratio)/t
k_1 = -np.log(ratio_1) / t_1
t_target_1 = -np.log(ratio_target) / k_1
extra_time_1 = t_target_1 - t_1

# --- Case b: Second Order (n=2) ---
# 1/CA - 1/CA0 = k*t => k*CA0 = (1/ratio - 1) / t
kCA0 = (1/ratio_1 - 1) / t_1
t_target_2 = (1/ratio_target - 1) / kCA0
extra_time_2 = t_target_2 - t_1

print(f"--- Problem A1 Results ---")
print(f"First Order (n=1): k = {k_1:.2f} h^-1")
print(f"Time to reach 0.001: {t_target_1:.2f} h (Extra: {extra_time_1:.2f} h)")
print(f"\nSecond Order (n=2): k*CA0 = {kCA0:.2f} h^-1")
print(f"Time to reach 0.001: {t_target_2:.2f} h (Extra: {extra_time_2:.2f} h)")
```

### 2. CSTR Design Comparison
For a CSTR at steady state with constant density:
$$\tau = \frac{C_{A0} - C_A}{-r_A}$$[cite: 2]

```python
# First Order CSTR: tau = (1/k) * ((1-ratio)/ratio)
tau_1 = (1 / k_1) * ((1 - ratio_target) / ratio_target)

# Second Order CSTR: tau = (1/kCA0) * ((1-ratio)/ratio^2)
tau_2 = (1 / kCA0) * ((1 - ratio_target) / (ratio_target**2))

print(f"--- Problem A2 Results ---")
print(f"CSTR Residence Time (n=1): {tau_1:.2f} hours")
print(f"CSTR Residence Time (n=2): {tau_2:,.2f} hours")
```

## Problem B: Gas Phase Batch Reactor
**Reaction:** $A + B \rightarrow C + D$ | **Rate:** $-r_A = k C_A C_B$

### b) Computing the Rate Constant k
We use the Ideal Gas Law: $C = \frac{P}{RT}$.
Initial pressure is 100 kPa with equal amounts of A and B, so $P_{A0} = 50$ kPa[cite: 2].

```python
R_gas = 8.3144 # Pa*m^3/(mol*K)
T1 = 25 + 273.15 # Kelvin
P_total = 100000 # Pa
ratio_B = 0.15
t_B = 1.0 # hour

# Initial concentration of A
CA0_B = (0.5 * P_total) / (R_gas * T1) #[cite: 2]

# Rate constant from Problem A integration
k_25C = (1/ratio_B - 1) / (CA0_B * t_B)
k_25C_sec = k_25C / 3600 # Convert to m^3/(mol*s)

# c) Temperature Dependence (Arrhenius) at 100C
T2 = 100 + 273.15
Ea_over_R = 10200 # K
k_100C = k_25C_sec * np.exp(Ea_over_R * (1/T1 - 1/T2))#[cite: 2]

print(f"--- Problem B Results ---")
print(f"Initial concentration CA0: {CA0_B:.2f} mol/m^3")
print(f"Rate constant k at 25C: {k_25C_sec:.3e} m^3/(mol*s)")
print(f"Rate constant k at 100C: {k_100C:.3e} m^3/(mol*s)")
```
